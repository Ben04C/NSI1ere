## D. Les dictionnaires

Si les tuples et les listes se ressemblent énormément, les **dictionnaires** en diffèrent nettement car leurs **éléments ne sont pas ordonnés**. On accède à la **valeur** d'un élément d'un dictionnaire par une **clé** qui peut être un nombre entier mais aussi une chaîne de caractère ou un nombre réel ...

Les dictionnaires sont notamment très utilisés pour contenir les **métadonnées d'un fichier**, comme par exemple les données EXIF d'une photo (appareil utilisé, localisation GPS, date de prise de vue, ...).

### I. Construire des dictionnaires

#### I.1 Créer un dictionnaire

Un dictionnaire s'écrit entre <span style="color:red;">**accolades**</span> et ses éléments sont séparés par des <span style="color:red;">**virgules**</span>.

Chaque élément s'écrit sous la forme `clé : valeur`. Il n'y a **qu'une seule valeur par clé** mais comme cette valeur peut être un *tuple*, une *liste* ou même un *dictionnaire*, ce n'est **pas limitant**.

Pour **ajouter élément**, on parle aussi d'une **entrée**, à un dictionnaire existant la syntaxe est `mon_dico[cle]=valeur` où `cle`est une clé qui n'existait pas auparavant.

```python
#Exemple 1 :
# création d’un dictionnaire vide
mon_dico={}
# création d’un dictionnaire pré-remplis
mon_dico={"écran": "screen","souris":"mouse"}
print(type(mon_dico)) # affichage du type
print(mon_dico) #affichage du dictionnaire
# ajout d’une entrée
mon_dico["clavier"]="keyboard"
print(mon_dico) #affichage du dictionnaire
mon_dico[2]=3
print(mon_dico) #affichage du dictionnaire
```

**Remarques :**

La fonction `dict()` permet également de convertir une liste de listes à 2 éléments en dictionnaire et on peut aussi **construire un dictionnaire par compréhension**.

```python
#Exemple 2:
liste=[['a',1],['b',2],['c',3]] #construction d'une liste de listes à 2 éléments
print(dict(liste)) #conversion en dictionnaire
d={x:x**2 for x in range(1,5)} #construction par compréhension d'un dictionnaire où x est la clé et x² la valeur
print(d)
```

#### I.2 Accéder aux éléments d'un dictionnaire

La méthode `.values()` renvoie la liste de **toutes les valeurs** d'un dictionnaire.

La méthode `.keys()` renvoie la liste de **toutes les clés** d'un dictionnaire.

La méthode `.items()` renvoie la liste de **toutes les éléments** d'un dictionnaire.

Pour **accèder à la valeur d'une clé**, la syntaxe est `mon_dico[cle]` mais on peut aussi utiliser la méthode `.get()` qui a l'avantage de ne pas renvoyer d'erreur si la clé demandée n'existe pas (elle renvoie alors `none`).

La fonction `len()` renvoie toujours la **longueur du dictionnaire** (le nombre de clés/entrées).

```python
#Exemple 3 :
print(mon_dico.values()) #Affiche les valeurs
print(mon_dico.keys()) #Affiche les clés
print(mon_dico.items()) #Affiche les éléments/entrées
print(mon_dico['souris']) #Affiche la valeur associée à la clé 'souris'
print(mon_dico.get('souris')) #Affiche la valeur associée à la clé 'souris'
```

### II. Manipuler des dictionnaires

Une fois construit, la **principale utilisation** d'un dictionnaire consiste simplement à **faire des recherches d'éléments** à l'intérieur. Mais on peut bien sûr aussi le modifier ou le copier.

#### II.1 Supprimer une entrée

L'instruction `del mon_dico[cle]` (ou `del(mon_dico[cle])`)permet de supprimer une entrée à partir de sa clé.

La méthode `.clear()` permet de vider entièrement un dictionnaire.


```python
#Exemple 4:
print(mon_dico)
del mon_dico['souris']
print(mon_dico)
```

#### II.2 Créer une copie d'un dictionnaire

La méthode `.copy()` et la fonction `dict()` permettent de créer une **copie distincte d'un dictionnaire**. Comme pour les listes l'instruction `mon_dico2=mon_dico` ne **crée pas un nouveau dictionnaire mais lui donne seulement un deuxième nom**. Par conséquent, toute modification ultérieure de `mon_dico` entrainera aussi la modification de `mon_dico2` et inversement.

**Remarque :**
Attention, si les valeurs du dictionnaire sont des listes, on retrouve ce même problème de deux noms différents pour les mêmes listes. Dans ce cas, on doit utiliser la fonction `deepcopy()` qui crée une **copie intégrale** (il faut l'importer avec `from copy import deepcopy`).

```python
#Exemple 5:
d=mon_dico.copy() # crée un nouveau dictionnaire identique au premier
print("mon_dico = ",mon_dico)
print("d = ",d) # les 2 dictionnaires sont bien identiques
d[4]=5 #on ajoute une entrée à d
print("d = ",d) #d est bien modifié ...
print("mon_dico = ",mon_dico) # ... mais pas mon_dico
mon_dico2=mon_dico #mon_dico2 est juste un autre nom de mon_dico
mon_dico['nouvelle entrée']='Test' #on ajoute une entrée à mon_dico
print("mon_dico = ", mon_dico)
print("mon_dico2 = ", mon_dico2)
```

#### II.3 Rechercher dans un dictionnaire

**Les dictionnaires sont implémentés pour optimiser la recherche d'un élément, de sorte que la durée d'une recherche dépend très peu de la longueur du dictionnaire.** C'est pourquoi ils sont à préférer aux listes si on sait que l'on va faire des recherches dans un grand nombre d'éléments.

L'instruction `in` permet de **rechercher la présence d'une valeur dans un objet itérable**. On l'a déjà rencontrée dans les boucles for comme `for i in range(n) :` ou `for el in liste :` par exemple.

C'est aussi elle qui permet de **chercher dans un dictionnaire** :
- pour **chercher une valeur** : l'expression `valeur in mon_dico.values()` vaut `True` si la valeur est présente dans le dictionnaire et `False` sinon.
- pour **chercher une clé** : l'expression `cle in mon_dico.keys()` vaut `True` si la clé est présente dans le dictionnaire et `False` sinon; l'expression `cle in mon_dico` est équivalente.


```python
#Exemple 6:
print(mon_dico)
#recherche d'une valeur
print("Y a-t-il la valeur 'mouse' dans le dictionnaire ?",'mouse' in mon_dico.values())
#Faux car 'mouse' n'apparaît nul part
print("Y a-t-il la valeur 'screen' dans le dictionnaire ?",'screen' in mon_dico.values())
#Vrai car 'screen' est la valeur associée à la clé 'écran'
print("Y a-t-il la valeur 'écran' dans le dictionnaire ?",'écran' in mon_dico.values())
#Faux car 'écran' est une clé, pas une valeur
#recherche d'une clé
print("Y a-t-il la clé 'mouse' dans le dictionnaire ?",'mouse' in mon_dico.keys(), 'mouse' in mon_dico)
#Faux car 'mouse' n'apparaît nul part
print("Y a-t-il la clé 'screen' dans le dictionnaire ?",'screen' in mon_dico.keys(),'screen' in mon_dico)
#Faux car 'screen' est une valeur, pas une clé
print("Y a-t-il la clé 'écran' dans le dictionnaire ?",'écran' in mon_dico.keys(), 'écran' in mon_dico)
#Vrai car 'écran' est la clé dont la valeur est 'screen'
```
