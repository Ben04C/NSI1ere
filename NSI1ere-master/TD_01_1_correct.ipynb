## 1. Conversion d'une base b vers la base décimale

### 1.1 Méthode de droite à gauche

Voici un algorithme permettant de convertir un nombre entier écrit en base $b$ vers la base décimale :
> resultat $\leftarrow$ 0<br>
> i $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de droite à gauche : <br>
>>    resultat $\leftarrow$ resultat + $c \times b^i$<br>
>>    i $\leftarrow$ i+1<br>
>
> Renvoyer resultat<br>

**Q1. Programmer une fonction ``convert_dec`` qui, selon l'algorithme précédent, prend en argument un nombre entier et la base dans laquelle il est écrit et renvoie sa valeur décimale. Penser à spécifier la fonction et éventuellement à utiliser ``assert``.**

```python
# Question 1
def convert_dec(nb_entier, b):
    """ nb_entier est de type int et positif; b est de type int et compris entre 2 et 16 inclus.
    Cette fonction renvoie sa valeur en base décimale de nb_entier écrit en base b."""
    assert type(b) == int, "La base indiquée doit être un nombre entier."
    assert (b>1) and (b<17), "La base indiquée doit être un entier compris entre 2 et 16."
    chaine=str(nb_entier) #permet d'accéder aux chiffres composant entier à l'aide d'indices
    long=len(chaine) #nombre de chiffres dans entier
    resultat=0
    for i in range(long): #i=0, 1, ... long-1
        if chaine[long-1-i] == ('A' or 'a') : #si ce "chiffre" est A alors il vaut 10
            resultat=resultat+10*b**i
        elif chaine[long-1-i] == ('B' or 'b') : #si ce "chiffre" est B alors il vaut 11
            resultat=resultat+11*b**i
        elif chaine[long-1-i] == ('C' or 'c') : #...
            resultat=resultat+12*b**i
        elif chaine[long-1-i] == ('D' or 'd') :
            resultat=resultat+13*b**i
        elif chaine[long-1-i] == ('E' or 'e') :
            resultat=resultat+14*b**i
        elif chaine[long-1-i] == ('F' or 'f') :
            resultat=resultat+15*b**i
        else : #sinon on applique la formule normale
            resultat=resultat+int(chaine[long-1-i])*b**i #on ajoute c*b**i où c est le ième chiffre de chaine en partant de droite
    return resultat
```

### 1.2 Méthode de gauche à droite (méthode de Horner)

Voici un 2ème algorihtme permettant aussi de convertir un nombre entier écrit en base $b$ vers la base décimale mais en parcourant le nombre dans l'autre sens :
> resultat $\leftarrow$ 0<br>
> **Pour** tous les chiffres $c$ allant de gauche à droite : <br>
>>    resultat $\leftarrow$ b $\times$ resultat<br>
>>    resultat $\leftarrow$ resultat + $c$<br>
>
> Renvoyer resultat<br>

**Remarque :** Cette méthode peut paraître plus complexe que la première mais elle est en fait bien plus efficace (rapide) car elle ne nécessite que des opérations simples (multiplication et addition) au lieu d'utiliser des puissances.

```python
# Question 3
def convert_horner(nb_entier, b):
    """ nb_entier est de type int et positif; b est de type int et compris entre 2 et 16 inclus.
    Cette fonction renvoie sa valeur en base décimale de nb_entier écrit en base b."""
    assert type(b) == int, "La base indiquée doit être un nombre entier."
    assert (b>1) and (b<17), "La base indiquée doit être un entier compris entre 2 et 16."
    chaine=str(nb_entier) #permet d'accéder aux chiffres composant entier à l'aide d'indices
    long=len(chaine) #nombre de chiffres dans entier
    resultat=0
    for i in range(long): #i=0, 1, ... long-1
        resultat=resultat*b
        if chaine[i] == ('A' or 'a') : #si ce "chiffre" est A alors il vaut 10
            resultat=resultat+10
        elif chaine[i] == ('B' or 'b') : #si ce "chiffre" est B alors il vaut 11
            resultat=resultat+11
        elif chaine[i] == ('C' or 'c') : #...
            resultat=resultat+12
        elif chaine[i] == ('D' or 'd') :
            resultat=resultat+13
        elif chaine[i] == ('E' or 'e') :
            resultat=resultat+14
        elif chaine[i] == ('F' or 'f') :
            resultat=resultat+15
        else : #sinon on applique la formule normale
            resultat=resultat+int(chaine[i]) #on ajoute c où c est le ième chiffre de chaine en partant de gauche
    return resultat
```

## 2. Conversion la base décimale vers une base b

### 2.1 Conversion par divisions successives par $b$
Pour convertir un nombre entier de la base décimale vers une base $b$, il suffit de le diviser par $b$ (division entière), de noter le reste puis de diviser à nouveau le résultat par $b$ et de noter le reste et ainsi de suite jusqu'à obtenir 0. En écrivant les restes obtenus en commençant par le dernier, on obtient le résultat de la conversion.

**Q5. Programmer une fonction ``convert_base`` qui, selon l'algorithme précédent, prend en argument un nombre entier dans la base décimale et la base dans laquelle on veut le convertir et renvoie la valeur convertie. Penser à spécifier la fonction et éventuellement à utiliser ``assert``.**

```python
# Question 5
def convert_base(nb_entier,b):
    """nb_entier est un entier positif et b un entier compris entre 2 et 16.
    Cette fonction renvoie la valeur de nb_entier dans la base b sous forme d'un entier si b<9 et
    d'une chaîne de caractères sinon. """
    assert type(b) == int, "La base indiquée doit être un nombre entier."
    assert (b>1) and (b<17), "La base indiquée doit être un entier compris entre 2 et 16."
    restes='' # chaîne de caractère vide
    quotient=nb_entier
    if nb_entier == 0 :
        return nb_entier
    while quotient != 0:
        temp=quotient%b #reste de la divions euclidienne par b
        if temp == 10 : # si le reste est supérieur à 9, on le remplace par une lettre
            temp = 'A'
        elif temp == 11 :
            temp = 'B'
        elif temp == 12 :
            temp = 'C'
        elif temp == 13 :
            temp = 'D'
        elif temp == 14 :
            temp = 'E'
        elif temp == 15 :
            temp = 'F'
        else :
            temp = str(quotient%b) # si le reste est < 10, on le convertit en chaîne de caractères
        restes=restes+temp #addtion de chaînes de caractères = concaténation
        quotient=quotient//b # le prochain quotient est obtenu en divisant par b
    resultat='' #il faut inverser les caractères pour le résultat
    for i in range(len(restes)):
        resultat=resultat+restes[len(restes)-i-1] #on ajoute les caractères de restes à resultat en commencant par la fin
    if b<10 :
        return int(resultat) # retourne un entier si la base < 10
    else :
        return resultat # retourne une chaîne de caractères sinon
```

**Q6. Utiliser la fonction``convert_base`` précédente pour convertir le nombre 213 dans les bases 2, 4 et 16.**

```python
# Question 6
print('Ecriture de 213 en base 2 :', convert_base(213,2))
print('Ecriture de 213 en base 4 :', convert_base(213,4))
print('Ecriture de 213 en base 16 :', convert_base(213,16))
```

### 2.2 Conversion par divisions successives par $b^i$
On utilisera plutôt cette méthode si on veut fixer le nombre $k$ de chiffres utilisés pour écrire le résultat, quitte à ce que ce résultat commence par des 0.

Il suffit alors d'effectuer la division entière du nombre à convertir par $b^{k-1}$ puis de recommencer en divisant le reste par $b^{k-2}$ et ainsi de suite jusqu'à diviser par $b^0=1$. Les résultats successifs de ces divisions dans l'ordre forment le nombre final. Evidemment, cela ne fonctionne pas si le nombre à convertir est trop grand pour le nombre de chiffres choisi pour le résultat.

Dans le cas d'une conversion d'un nombre entier $n$ en nombre binaire de k chiffres, l'algorithme est le suivant :
> **Pour** $i$ allant de $k-1$ à $0$ : <br>
>> **Si** $n \geq 2^i$ :<br>
>>>    $n \leftarrow n-2^i $<br>
>>>    Rajouter le chiffre $1$ au résultat<br>
>>
>> **Sinon** :<br>
>>> Rajouter le chiffre $0$ au résultat<br>
>
> Renvoyer resultat<br>

**Q7. Programmer une fonction ``convert_bin`` qui, selon l'algorithme précédent, prend en argument un nombre entier dans la base décimale et le nombre de chiffres souhaité pour le résultat et renvoie la valeur convertie en binaire. Penser à spécifier la fonction et éventuellement à utiliser ``assert``.**

```python
# Question 7
def convert_bin(nb_entier,k):
    """nb_entier est un entier positif en base décimale et k un entier > 0.
    Cette fonction renvoie une chaine de caractères contenant la valeur de nb_entier convertie en un nombre binaire de k chiffres. """
    assert (type(k) == int) and (k>0), "Le nombre de chiffres du résultat doit être un entier > 0."
    assert (type(nb_entier) == int) and (nb_entier >0), "Le nombre à convertir doit être un entier positif."
    assert nb_entier<2**k, "Le nombre à convertir est trop grand pour écrit avec aussi peu de chiffres."
    resultat='' #chaîne vide
    for i in range(k-1,-1,-1): #i=k-1, k-2, ... 1, 0
        if nb_entier >= 2**i : # si nb_entier > ou = à 2^i, on ajoute 1 à resultat et retire 2^i à nb_entier
            nb_entier=nb_entier-2**i
            resultat=resultat+'1'
        else : #sinon on rajoute un 0 à resultat
            resultat=resultat+'0'
    return resultat #si on retourne int(resultat), les 0 situés à gauche sont supprimés
```

```python
# Question 8
print("En binaire sur 8 chiffres le nombre 197 s'écrit :", convert_bin(197,8)," soit ", bin(197))
print("En binaire sur 8 chiffres le nombre 63 s'écrit :", convert_bin(63,8)," soit ", bin(63))
print("En binaire sur 8 chiffres le nombre 78 s'écrit :", convert_bin(78,8)," soit ", bin(78))
```
