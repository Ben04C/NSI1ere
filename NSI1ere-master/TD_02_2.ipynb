**Q1. Écrire une fonction qui prend en paramètre une liste d’entiers naturels et qui renvoie la somme des éléments de la liste.**

```python
#Question 1 :
def aze(L1):
    assert type(L1)==list, "il faut que L1 soit une liste"
    result=0
    for num in L1:
        result+=num
    return result

aze([1,5])
```

**Q4. Écrire une fonction qui prend en paramètre un entier naturel et qui renvoie la liste des nombres pairs inférieurs à cet entier.**

```python
#Question 4 :
def infpairs(N4):
    L4=[]
    for i in range (N4-1, 0, -1):
        if i%2==0:
            L4.append(i)
    return L4
print(infpairs(50))
```

**Q7. Écrire une fonction qui prend en paramètre une liste d'entiers naturels et qui renvoie une liste constituée des doubles de celle reçue en paramètre.**

```python
#Question 7 :
def double(L7):
    assert type(L7)==list, "il faut que L1 soit une liste"
    result7=[]
    for num in L7:
        result7.append(num*2)
    return result7
print(double([2,4,6]))
```

**Q8. Étant donné une liste d’entiers positifs et négatifs, écrire une fonction qui place en début de liste les éléments négatifs et en fin de liste les éléments positifs.**

On utilisera deux listes intermédiaires `Lneg` et `Lpos` qui contiendront les éléments positifs et négatifs de la liste de départ.
La fonction retournera  `Lneg + Lpos`.

```python
#Question 8 :
def seperate(L8):
    assert type(L8)==list, "il faut que L1 soit une liste"
    Lneg=[]
    Lpos=[]
    for num in L8:
        if num >= 0:
            Lpos.append(num)
        else:
            Lneg.append(num)
    return Lneg+Lpos
print(seperate([-1,5,-5,7,6,-8]))
```
