## Tâche 1 :


Écrire un programme qui calcule le nombre de survivants.

**Attention :**

* La valeur de la clé 'Survived' est de type `string`.

* Pour l'atteindre sa valeur, on peut utiliser l'instruction : `titanic[n° de ligne]['Survived']`


```python
#executer ce scripte en premier afin de créer une liste "titanic" avec les données du fichier csv titanic.csv
import csv

reader = csv.DictReader(open('Files/titanic.csv', 'r'))
titanic = []
for row in reader:
    titanic.append(dict(row))
```

```python
# Écrire le script ici
survived=0
for info in titanic:
    if info["Survived"]=="1":
        survived+=1
print("Il y a eu", survived, "survivants")
```

## Tâche 2 :
Écrire un programme qui affiche le pourcentage de survivants par classe.

```python
# Écrire le script ici
#création de liste qui vont contenir tous les passagers de chaque classe.
class1=[]
class2 = []
class3 = []
#survived est une variable où il y aura le nombre de survivants dans chaque classe. Total est le nombre total de passagers de chaque classe.
survived1=0
total1=0
survived2 = 0
total2 = 0
survived3 = 0
total3 = 0

#on vérifie la classe de chaque passager et....
for info in titanic:
    if info["Pclass"]=="1": #...on l'ajouté à la liste class1
        class1.append(info)
    elif info["Pclass"]=="2": #...on l'ajouté à la liste class2
        class2.append(info)
    elif info["Pclass"]=="3": #...on l'ajouté à la liste class3
        class3.append(info)
        
#on vérifie le si les passagers ont survécu

for info in class1:
    if info["Survived"]=="1":#En classe1
        survived1+=1
    total1+=1
for info in class2:
    if info["Survived"]=="1":#En classe2
        survived2+=1
    total2+=1
for info in class3:
    if info["Survived"]=="1":#En classe 3
        survived3+=1
    total3+=1
#on calcule le pourcentage de chaque classe.

pourcentage1=survived1/total1*100
poucentage2=survived2/total2*100
pourcentage3=survived3/total3*100
#On affiche les résultats.
print("Il y a eu",int(pourcentage1),"% de survivants en classe1")
print("Il y a eu",int(poucentage2),"% de survivants en classe2")
print("Il y a eu",int(pourcentage3),"% de survivants en classe3")
```

## Tâche 3  :
Écrire un programme qui donne le nombre de survivants embarqués à Cherbourg.

```python
# Écrire le script ici
#création d'une variable où l'on ajoute le nombre de survivants de Cherbourg
Csurvived=0
for info in titanic:
    if info["Embarked"]=="C": #vérifie si le passager à embarqué a Cherbourg
        if info["Survived"]=="1": #vérifie si le passager (on a déjà vérifié s'il vennait de Cherbourg) à survécu.
            Csurvived+=1
print("Il y a eu",Csurvived,"survivants qui ont embarqué à Cherbourg")
```
